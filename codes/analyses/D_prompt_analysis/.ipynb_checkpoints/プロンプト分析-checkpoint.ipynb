```python
import csv
import torch
from tokenizers import Tokenizer
from transformers import RobertaConfig, RobertaForMaskedLM, PreTrainedTokenizerFast
import random
import numpy as np
from tqdm import tqdm  # 進捗表示のため

import os
import csv


# GPUの設定
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# ボキャブラリーの読み込み
#fin = '/Users/yuikoshitak/Documents/GraduateSchool/Master_Research/prj-evo_design/model_parameter/vocab_pfam_product_hmmer33.json'
fin = '/working.3/Kawano/vocab_pfam_product_hmmer33_revised.json'

# トークナイザーを定義
tokenizer_obj = Tokenizer.from_file(fin)
tokenizer_obj.add_special_tokens(["MASK"])
tokenizer_obj.add_special_tokens(["[PAD]"])
tokenizer = PreTrainedTokenizerFast(tokenizer_object=tokenizer_obj)
tokenizer.mask_token = "[MASK]"
tokenizer.pad_token = "[PAD]"
tokenizer.sep_token = "[SEP]"

# RoBERTaのモデルの設定
config = RobertaConfig(
    vocab_size=19723,
    max_position_embeddings=514,
    hidden_size=1024,
    num_attention_heads=16,
    num_hidden_layers=8,
    type_vocab_size=1
)
model = RobertaForMaskedLM(config).to(device)

# 学習済みパラメータのロード
#model.load_state_dict(torch.load('/Users/yuikoshitak/Documents/GraduateSchool/Master_Research/prj-evo_design/model_parameter/fungi/Model4_20240918_epoch150.pth', map_location=torch.device('mps')))
model.load_state_dict(torch.load('/working.3/Kawano/model-parameter/RoBERTa_bacteria+fungi20240918_epoch150.pth', map_location=torch.device('cuda')))



# 入力ファイル
input_file = "Modified_Sorted_MIBiG_domain_list.txt"

# トークナイザーの全語彙を取得
vocab = list(tokenizer.get_vocab().keys())

# トークン数の範囲指定
token_count_ranges = list(range(6, 15))  # 6個から15個

# 出力ディレクトリ
output_dir = "output_files_20250214"
os.makedirs(output_dir, exist_ok=True)

# テキストファイルを一行ずつ処理
with open(input_file, 'r', encoding='utf-8') as f:
    lines = f.readlines()

for token_count in token_count_ranges:
    all_results = []  # 各トークン数範囲ごとの結果を格納

    for line in tqdm(lines, desc=f"Processing lines with {token_count} tokens"):
        line = line.strip()
        if not line:
            continue

        # 行のトークンを分割し、3単語目以降を取得
        tokens = line.split()[2:]
        bgc_id = line.split()[0]

        # [SEP]を除いたトークンを取得
        candidates = [token for token in tokens if token != "[SEP]"]

        if not (token_count == len(candidates)):
            continue  # トークン数が指定範囲内でなければスキップ

        max_probability = -1
        target_token = None

        for idx in range(len(tokens)):
            token = tokens[idx]
            masked_tokens = tokens.copy()
            if masked_tokens[idx] == "[SEP]":
                continue
            #print(masked_tokens[idx])
            masked_tokens[idx] = "[MASK]"
            
            tokens_id = tokenizer.convert_tokens_to_ids(masked_tokens)
            tokens_tensor = torch.tensor([tokens_id]).to(device)

            with torch.no_grad():
                outputs = model(tokens_tensor)
                predictions = outputs[0]

            # MASK位置の予測スコアを取得
            softmax_scores = torch.softmax(predictions[0, idx], dim=-1)
            token_id = tokenizer.convert_tokens_to_ids([token])[0]

            if softmax_scores[token_id].item() > max_probability:
                max_probability = softmax_scores[token_id].item()
                target_token = token
                target_idx = idx

        # ターゲットトークンが見つからない場合はスキップ
        if target_token is None:
            continue

        # 行の結果を保存するリスト
        line_results = [bgc_id, target_token]

        # ランダムに置き換える数を0から(token_count)個まで変化させる
        for num_replacements in range(0, token_count):
            avg_probabilities = []

            # 10回のランダム置き換え
            for _ in range(10):
                # トークンをコピーして操作
                modified_tokens = tokens.copy()

                # トークンを置き換える場合
                if num_replacements > 0:
                    available_indices = [i for i in range(len(modified_tokens)) 
                         if i != target_idx and modified_tokens[i] != "[SEP]"]  # target_idx と[SEP]のインデックスを除外
                    replace_indices = random.sample(available_indices, num_replacements)

                    # トークナイザーの語彙からランダムにサンプリングして置き換え
                    for idx in replace_indices:
                        sampled_token = random.choice(vocab)
                        modified_tokens[idx] = sampled_token

                # ターゲットトークンを MASK
                masked_tokens = modified_tokens.copy()
                
                masked_tokens[target_idx] = "[MASK]"

                # トークン列をテンソルに変換
                tokens_id = tokenizer.convert_tokens_to_ids(masked_tokens)
                tokens_tensor = torch.tensor([tokens_id]).to(device)

                # モデルに入力して予測
                with torch.no_grad():
                    outputs = model(tokens_tensor)
                    predictions = outputs[0]

                # MASKの位置の予測スコアを取得
                softmax_scores = torch.softmax(predictions[0, target_idx], dim=-1)
                target_token_id = tokenizer.convert_tokens_to_ids([target_token])[0]
                avg_probabilities.append(softmax_scores[target_token_id].item())

            # 平均確率を計算
            mean_probability = np.mean(avg_probabilities)
            line_results.append(mean_probability)

        # 行ごとの結果を追加
        all_results.append(line_results)

    # 結果をCSVに保存
    output_file = os.path.join(output_dir, f"token_prediction_{token_count}_tokens.csv")
    with open(output_file, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)

        # ヘッダー行を書き込む
        headers = ["BGC_ID", "masked_token"] + [f"Num_Replacements={i}" for i in range(token_count)]
        writer.writerow(headers)

        # 各行の結果を書き込む
        writer.writerows(all_results)

print("Done")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

# ディレクトリ内のCSVファイルを一括処理する関数
def process_csv_files(directory_path):
    # 出力用ディレクトリ
    output_directory = os.path.join(directory_path, "output_plots")
    os.makedirs(output_directory, exist_ok=True)  # ディレクトリがない場合は作成
    
    # ディレクトリ内の全てのCSVファイルを取得
    csv_files = [f for f in os.listdir(directory_path) if f.endswith('.csv')]
    
    for file_name in csv_files:
        file_path = os.path.join(directory_path, file_name)
        num = file_name.split("_")[2]
        # CSVファイルの読み込み
        data = pd.read_csv(file_path)

        # "Num_Replacements="の列を動的に取得
        num_columns = [col for col in data.columns if col.startswith("Num_Replacements=")]
        
        # 数値データとして変換
        numeric_data = data[num_columns].astype(float)
        
        # 平均値と標準偏差の計算
        mean_values = numeric_data.mean()
        std_values = numeric_data.std()

        # NumPy配列に変換
        mean_values_np = mean_values.to_numpy()
        std_values_np = std_values.to_numpy()

        # 横軸の値（0～最大Num_Replacements値）
        x = np.array(range(len(num_columns)))

        # プロット
        plt.figure(figsize=(10, 6))
        plt.plot(x, mean_values_np, label="Mean", marker='o')
        plt.fill_between(x, mean_values_np - std_values_np, mean_values_np + std_values_np, color='gray', alpha=0.3, label="±1 Std Dev")

        # グラフの設定
        plt.xlabel('Num_Replacements')
        plt.ylabel('Values')
        plt.title(f'Prediction Accuracy Declines with Increased Word Replacements Using {num}-domains from BGC')
        plt.xticks(x, [col.split('=')[1] for col in num_columns])  # "Num_Replacements=X"のXを表示
        plt.legend(prop={'size': 20})
        plt.grid(True)

        # 目盛りのラベルサイズを調整
        plt.tick_params(axis='both', which='major', labelsize=14)  # 主要目盛りのサイズを14に設定
        plt.tick_params(axis='both', which='minor', labelsize=12)  # 補助目盛りのサイズを12に設定（必要なら）

        plt.tight_layout()

        # 画像ファイルとして保存
        output_path = os.path.join(output_directory, f"{file_name.replace('.csv', '.png')}")
        plt.savefig(output_path)
        print(f"Saved plot to {output_path}")

        # グラフを表示
        plt.show()

# ディレクトリパスの指定
directory_path = 'output_files_20250214/'

# 関数実行
process_csv_files(directory_path)
```
